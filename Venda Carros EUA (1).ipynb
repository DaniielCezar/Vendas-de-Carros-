### Agora que temos tudo em mãos, vamos traduzir os nomes das colunas. 

```python
base.columns = ['Ano', 'Marca', 'Modelo', 'Usado/Novo', 'Preço', 'Classificação do Consumidor', 'Avaliações do consumidor', 
              'Tipo de vendedor', 'Nome do Vendedor', 'Avaliação do vendendor', 'Comentarios do vendedor', 'Rua do Vendedor', 'Estado', 'CEP', 'Qualidade do Negocio',
             'Conforto', 'Desing do Interior', 'Desempenho', 'Preço Avaliado pelos consumidores', 'Desing Exterior',
             'Confiabilidade', 'Cor do Exterior', 'Cor do Interior','Drivetrain', 'MinMPG', 'MaxMPG', 'Tipo de Combustivel', 
                'Transmissão', 'Motor', 'VIN', 'Stock#', 'Milhagem']

```

```python
base
```

### Seguindo a mesma ideia, vamos para a tradução dos dados. 

```python
base['Usado/Novo']= base['Usado/Novo'].replace(['Used'],'Usado')

base['Tipo de vendedor']= base['Tipo de vendedor'].replace(['Dealer'],'Distribuidor')

base['Tipo de vendedor']= base['Tipo de vendedor'].replace(['Private'],'Privado')

base['Qualidade do Negocio']= base['Qualidade do Negocio'].replace(['Great'],'Ótimo')

base['Qualidade do Negocio']= base['Qualidade do Negocio'].replace(['Good'],'Bom')

base['Qualidade do Negocio']= base['Qualidade do Negocio'].replace(['Fair'],'Razoável')

base
```

### Indo para o proximo passo, vamos verificar quais dados estão faltando. 

```python

faltantes = base.isnull().sum().sort_values(ascending=False)

faltantes
```

### Como verificado acima, apenas a coluna "Qualidade do Negocio" tem dados faltantes. 

### Então iremos trabalhar eles da seguinte forma: Substituindo os dados faltantes para a informação "Sem Resposta", assim não perdemos os demais conteudos da linha. 

```python
base['Qualidade do Negocio'].fillna('Sem Resposta', inplace = True)

base
```

### Agora vamos verificar o tipo dos dados de cada coluna. 

```python
base.dtypes
```

### Como visto acima a coluna CEP esta como objeto pois foi inserido algumas strings em determinadas linhas 

### Com a ajuda do Power BI conseguimos indentiifica-los e vamoss excluir elas, e em seguida transformar o tipo de dado da coluna em "int"

```python
base = base.drop([689,1424,1985,3034,5126,6189,6269,7003,7265,8110])

base
```

```python
base['CEP'] = base['CEP'].astype('int64')
```

### Verificando se a transformação deu certo e observando a coluna Preço, vemos que ela támbem esta como objeto, isso devido ao "$" que está inserido antes do valor.

### Então vamos excluir o "$" e transformar o tipo da coluna como "int".

```python
base.dtypes
```

```python
#Remove rows with 'Not Priced'
base = base[base['Preço'] != 'Not Priced']

#Remove symbols and convert all entries to integers
base['Preço'] = [Preço.replace('$','') for Preço in base['Preço']]
base['Preço'] = [int(Preço.replace(',','')) for Preço in base['Preço']]

base
```

```python
base.dtypes
```

### Agora separando as informações do da coluna "Usado/Novo" vemos que todo veiculo novo está como "Certified" + o nome da marca.

### Juntando todas as linhas que tem a infomação "certified" e mudando para a palavra "Novo", em seguida verificamos se deu certo vendo as linhas 9 e 40.

```python
base['Usado/Novo'] = ['Novo' if 'certified' in item.casefold()
                  else 'Usado' for item in base['Usado/Novo']]
```
